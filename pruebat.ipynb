```python
def describir_tendencias(df_pred):
    resumen = []
    ult_ano = df_pred['Year'].max()

    for pais in df_pred['Country'].unique():
        datos = df_pred[df_pred['Country'] == pais].sort_values('Year')
        if datos.shape[0] < 5:
            continue
        inicial = datos['yhat'].iloc[0]
        final = datos['yhat'].iloc[-1]

        variacion = final - inicial
        tendencia = "aumentÃ³" if variacion > 0 else "disminuyÃ³" if variacion < 0 else "se mantuvo estable"
        porcentaje = abs(variacion / inicial) * 100 if inicial else 0

        resumen.append(f"ðŸ”¹ En {pais}, el consumo proyectado {tendencia} de {int(inicial)} a {int(final)} entre {datos['Year'].iloc[0]} y {ult_ano} ({porcentaje:.1f}% de cambio).")

    if resumen:
        display(HTML("<h4>ðŸ§  DescripciÃ³n automÃ¡tica del grÃ¡fico</h4>"))
        for frase in resumen:
            display(HTML(f"<p>{frase}</p>"))
    else:
        print("No se pudo generar descripciÃ³n automÃ¡tica.")
```

```python
def actualizar_vista(change):
    with output:
        clear_output()
        tipo = change['new']

        # Filtrar por tipo
        df_tipo = df_long[df_long['Coffee type'] == tipo]

        # Agrupar por aÃ±o y paÃ­s (consumo real)
        datos = df_tipo.groupby(['Year', 'Country'])['Consumption'].sum().reset_index()

        # Preparar tabla dinÃ¡mica
        tabla = datos.pivot(index='Year', columns='Country', values='Consumption').fillna(0).round(2)

        # Mostrar tabla original
        display(HTML(f"<h4>ðŸ“Š Consumo anual por paÃ­s - {tipo}</h4>"))
        display(tabla)

        # Preparar proyecciones hasta 2040
        paises = tabla.columns.tolist()
        aÃ±os_reales = tabla.index.tolist()
        ult_aÃ±o = max(aÃ±os_reales)
        aÃ±os_extra = 2040 - ult_aÃ±o

        # DF para almacenar predicciones
        df_pred = pd.DataFrame()

        for pais in paises:
            serie = tabla[pais]
            if serie.sum() == 0:
                continue
            df_model = pd.DataFrame({'ds': pd.to_datetime(serie.index, format='%Y'), 'y': serie.values})
            try:
                modelo = Prophet()
                modelo.fit(df_model)
                future = modelo.make_future_dataframe(periods=aÃ±os_extra, freq='Y')
                forecast = modelo.predict(future)
                forecast['Country'] = pais
                df_pred = pd.concat([df_pred, forecast[['ds', 'yhat', 'Country']]], axis=0)
            except:
                continue

        if df_pred.empty:
            print("âš ï¸ No se pudieron generar proyecciones.")
            return

        # Preparar grÃ¡fico
        df_pred['Year'] = df_pred['ds'].dt.year
        df_plot = df_pred[df_pred['Year'] <= 2040]
        df_plot = df_plot[df_plot['Year'] >= min(tabla.index)]  # limitar desde el primer aÃ±o real

        plt.figure(figsize=(14, 6))
        for pais in df_plot['Country'].unique():
            datos_pais = df_plot[df_plot['Country'] == pais]
            plt.plot(datos_pais['Year'], datos_pais['yhat'], label=pais)

        plt.title(f"ðŸ“ˆ Consumo real y proyectado por paÃ­s - {tipo} (hasta 2040)", fontsize=14)
        plt.xlabel("AÃ±o")
        plt.ylabel("Consumo")
        plt.legend(title="PaÃ­s", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        # Analisis de tendencias
        describir_tendencias(df_plot)

        # Mostrar tabla resumida final
        tabla_pred = df_plot.pivot(index='Year', columns='Country', values='yhat').round(2)
        display(HTML("<h4>ðŸ”® Tabla de consumo proyectado por paÃ­s (hasta 2040)</h4>"))
        display(tabla_pred)
```
