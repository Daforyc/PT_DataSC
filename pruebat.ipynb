```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from prophet import Prophet
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
```

```python
import warnings
import logging
```

```python
# Ocultar mensajes Prophet/cmdstanpy
logging.getLogger('cmdstanpy').disabled = True

# Silenciar advertencias
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)

# Fuente compatible con emojis (si usas emojis en títulos de gráficos)
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
# Limpieza previa
df_long['Country'] = df_long['Country'].str.strip()
df_long['Coffee type'] = df_long['Coffee type'].str.strip()
df_long = df_long.dropna(subset=['Consumption'])
```

```python
# Widget para tipo de café
coffee_selector = widgets.Dropdown(
    options=sorted(df_long['Coffee type'].unique()),
    description='☕ Tipo de café:'
)

output = widgets.Output()
```

```python
def describir_tendencias(df_pred):
    resumen = []
    ult_ano = df_pred['Year'].max()

    for pais in df_pred['Country'].unique():
        datos = df_pred[df_pred['Country'] == pais].sort_values('Year')
        if datos.shape[0] < 5:
            continue
        inicial = datos['yhat'].iloc[0]
        final = datos['yhat'].iloc[-1]

        variacion = final - inicial
        tendencia = "aumentó" if variacion > 0 else "disminuyó" if variacion < 0 else "se mantuvo estable"
        porcentaje = abs(variacion / inicial) * 100 if inicial else 0

        resumen.append(f"🔹 En {pais}, el consumo proyectado {tendencia} de {int(inicial)} a {int(final)} entre {datos['Year'].iloc[0]} y {ult_ano} ({porcentaje:.1f}% de cambio).")

    if resumen:
        display(HTML("<h4>🧠 Descripción automática del gráfico</h4>"))
        for frase in resumen:
            display(HTML(f"<p>{frase}</p>"))
    else:
        print("No se pudo generar descripción automática.")
```

```python
def actualizar_vista(change):
    with output:
        clear_output()
        tipo = change['new']

        # Filtrar por tipo
        df_tipo = df_long[df_long['Coffee type'] == tipo]

        # Agrupar por año y país (consumo real)
        datos = df_tipo.groupby(['Year', 'Country'])['Consumption'].sum().reset_index()

        # Preparar tabla dinámica
        tabla = datos.pivot(index='Year', columns='Country', values='Consumption').fillna(0).round(2)

        # Mostrar tabla original
        display(HTML(f"<h4>📊 Consumo anual por país - {tipo}</h4>"))
        display(tabla)

        # Preparar proyecciones hasta 2040
        paises = tabla.columns.tolist()
        años_reales = tabla.index.tolist()
        ult_año = max(años_reales)
        años_extra = 2040 - ult_año

        # DF para almacenar predicciones
        df_pred = pd.DataFrame()

        for pais in paises:
            serie = tabla[pais]
            if serie.sum() == 0:
                continue
            df_model = pd.DataFrame({'ds': pd.to_datetime(serie.index, format='%Y'), 'y': serie.values})
            try:
                modelo = Prophet()
                modelo.fit(df_model)
                future = modelo.make_future_dataframe(periods=años_extra, freq='Y')
                forecast = modelo.predict(future)
                forecast['Country'] = pais
                df_pred = pd.concat([df_pred, forecast[['ds', 'yhat', 'Country']]], axis=0)
            except:
                continue

        if df_pred.empty:
            print("⚠️ No se pudieron generar proyecciones.")
            return

        # Preparar gráfico
        df_pred['Year'] = df_pred['ds'].dt.year
        df_plot = df_pred[df_pred['Year'] <= 2040]
        df_plot = df_plot[df_plot['Year'] >= min(tabla.index)]  # limitar desde el primer año real

        plt.figure(figsize=(14, 6))
        for pais in df_plot['Country'].unique():
            datos_pais = df_plot[df_plot['Country'] == pais]
            plt.plot(datos_pais['Year'], datos_pais['yhat'], label=pais)

        plt.title(f"📈 Consumo real y proyectado por país - {tipo} (hasta 2040)", fontsize=14)
        plt.xlabel("Año")
        plt.ylabel("Consumo")
        plt.legend(title="País", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        # Analisis de tendencias
        describir_tendencias(df_plot)

        # Mostrar tabla resumida final
        tabla_pred = df_plot.pivot(index='Year', columns='Country', values='yhat').round(2)
        display(HTML("<h4>🔮 Tabla de consumo proyectado por país (hasta 2040)</h4>"))
        display(tabla_pred)
```

```python
# Conectar selector
coffee_selector.observe(actualizar_vista, names='value')
```

```python
# Mostrar interfaz
display(coffee_selector, output)
```

```python
# Lanzar con selección inicial
actualizar_vista({'new': coffee_selector.value})
```
